```python
#Inheritance
'''
    inheritance in python is a feature that allows a child class
    to inherit the properties and behaviour(methods and attributes)
    of another class(parent class).
    Types  of inheritance:
    single inheritance
    multiple inheritance
    Multilevel inheritance
    Hierarchical Inheritance
    Hybrid inheritance
'''

```

```python

#Single Inheritance
'''
    '-------------
    'Parent Class'
    '------------'
           ^           #which contain only one parent and only one child class
           '                
           '
    '-------------
    'Child Class '
    '------------'
           
'''
class Animal:
    def __init__(self,name):
        self.name=name
        self.is_alive=True
    def eat(self):
        print(f"{self.name} is eating")
    def sleep(self):
        print(f"{self.name} is sleeping")

class Dog(Animal):
    pass

    def bark(self):
        print("Dog barks")

dog =Dog("Puntu")
print(dog.name)
dog.eat()
dog.sleep()
dog.bark()
```

```python

#Multilevel inheritance
'''
  '-------------
    'Parent Class'
    '------------'
           ^           #which contain only one parent and only Multiple child class
           '                
           '
    '-------------
    'Child Class1'
    '------------'
           ^           
           '                
           '
    '-------------
    'Child Class2'
    '------------'

'''
class Shape:
    def __init__(self,name):
        self.name=name
    def display_name(self):
        print(f"Shape:{self.name}")
class rectangle(Shape):
    def __init__(self,name,length,breadth):
        super().__init__(name)
        self.length=length
        self.breadth=breadth

    def area(self):
        return self.length*self.breadth

class cuboid(rectangle):
    def __init__(self,name,length,breadth,height):
        super().__init__(name,length,breadth)
        self.height=height
    def volume(self):
        return self.length*self.breadth*self.height
obj=Shape("General Shape")
obj.display_name()

rect=rectangle("Rectangele",10,20)
rect.display_name()
print(f"Area of {rect.name}is{rect.area()}")

cuboid = cuboid("Cuboid", 10, 5, 7)
cuboid.display_name()
print(f"Area of base {cuboid.name}: {cuboid.area()}")
print(f"Volume of {cuboid.name}: {cuboid.volume()}")
```

```python
#Multiple Inheritance
'''
    '-------------      '------------'         '-------------'         
    'Parent Class'      'Parent Class2'        'Parent Class3'
    '------------'      '-------------'        '--------------'
          ^           #which ^contain many parent and ^only  child class
           '                 '                        '
             '               '                       '
               '             '                     '
                 ' ' ' ' '   '    ' ' ' ' ' ' ' ' '
                         '-------------
                         'Child Class1'
                         '------------'
'''
#parent class1 :Amish (software Developer)
class Amish:
    ProgrammingLanguage="Kotlin"
    def software_developer(self):
        print("Software is developed using",self.ProgrammingLanguage)
#parent class2 :Abeeral(Data Scientist)
class Abeeral:
    ProgramingLanguage="Python"
    def data_Scientist(self):
        print("Data Analysis and models are Built using ",self.ProgramingLanguage)

#child class :Kshitz (Business man with Multiple role)
class Kshitz(Amish,Abeeral):
    Brand_name="Ks_mandu"
    def Business(self):
        print(f"Kshitz runs a business name{self.Brand_name}")
    def display_roles(self):
        print("Kshitz is an Entrepreneur in the technology sector")
obj = Kshitz()
obj.software_developer()
obj.data_Scientist()
obj.Business()
obj.display_roles()
#output
'''
Software is developed using Kotlin
Data Analysis and models are Built using  Python
Kshitz runs a business nameKs_mandu
Kshitz is an Entrepreneur in the technology sector
'''
```

```python
#Hierarchical  inheritance
'''
    inheritance which contain only one parent class but multiple child class but 
    each child class can access parent class property
'''
import math
#parent class
class math_operations:
    def __init__(self,number):
        self.number=number
    def display_number(self):
        print(f"Number is:{self.number}")

#child class 1
class fact(math_operations):
    def calc_fact(self):
        print(f"The factorial of {self.number} is :{math.factorial(self.number)}")
#child class 2
class fibonacci(math_operations):
    def generate_fib(self):
        fb=[0,1]
        for _ in range(2,self.number):
            fb.append(fb[-1]+fb[-2])
        print(f"fibonacci series for first {self.number}terms is{fb[:self.number]}")

class Quadratic_solver(math_operations):
    def solve_quad(self,a,b,c):
        discrimant=b**2-4*a*c
        if discrimant>=0:
            root1=(-b+math.sqrt(discrimant))/(2*a)
            root2=(-b-math.sqrt(discrimant))/(2*a)
            print(f"Roots of quadratic equation are {root1},{root2}")
        else:
            print("No real root exist!!")

factorial = fact(5)
factorial.display_number()
factorial.calc_fact()

fibonacci = fibonacci(7)
fibonacci.display_number()
fibonacci.generate_fib()

solver = Quadratic_solver(0)
solver.solve_quad(1, -3, 2)
```

```python
#Hybrid inheritance
'''
    In Hybrid inheritance multiple inheritance and multilevel inheritance are 
    are combined in a single program.
'''
# Parent class 1: Circle
class Circle:
    def validate_circle(self, x, y, radius):
        equation = x**2 + y**2
        print(f"Point ({x}, {y}) in circle equation: x² + y² = r²")
        if equation == radius**2:
            print(f"The point lies on the circle (r = {radius}).")
        elif equation < radius**2:
            print(f"The point lies inside the circle (r = {radius}).")
        else:
            print(f"The point lies outside the circle (r = {radius}).")

# Parent class 2: Ellipse
class Ellipse:
    def validate_ellipse(self, x, y, a, b):
        equation = (x**2 / a**2) + (y**2 / b**2)
        print(f"Point ({x}, {y}) in ellipse equation: (x²/a²) + (y²/b²) = 1")
        if round(equation, 2) == 1:
            print(f"The point lies on the ellipse (a = {a}, b = {b}).")
        elif equation < 1:
            print(f"The point lies inside the ellipse (a = {a}, b = {b}).")
        else:
            print(f"The point lies outside the ellipse (a = {a}, b = {b}).")

# Child class: GeometryValidator (inherits from Circle and Ellipse)
class GeometryValidator(Circle, Ellipse):
    def validate_both(self, x, y, radius, a, b):
        print("\nValidating point for both circle and ellipse:")
        self.validate_circle(x, y, radius)
        self.validate_ellipse(x, y, a, b)

# Usage
validator = GeometryValidator()
validator.validate_circle(3, 4, 5)
validator.validate_ellipse(2, 3, 4, 5)
validator.validate_both(3, 4, 5, 6, 8)
```
